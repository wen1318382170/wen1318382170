```python
#导包
import pandas as pd
import numpy as np
from pandas import read_csv
import requests
from pandas.core.frame import DataFrame
import json
```

```python
gp=pd.read_csv(r'./数据/股票型前三.csv',dtype={'代码':str})#让代码读取数据为str
print(gp)
hh1=pd.read_csv(r'./数据/混合型-灵活前三.csv',dtype={'代码':str})#让代码读取数据为str
print(hh1)
hh2=pd.read_csv(r'./数据/混合型-偏股前三.csv',dtype={'代码':str})#让代码读取数据为str
print(hh2)
zs=pd.read_csv(r'./数据/指数型-股票前三.csv',dtype={'代码':str})#让代码读取数据为str
print(zs)
```

```python
q=pd.concat([gp, hh1,hh2,zs])
q['代码']
```

```python
#创建列表保存数据
code_l=[]#代码
time=[]#时间
per=[]#涨幅
nav_l=[]#净值

for i in q['代码']:
    url = "https://danjuanapp.com/djapi/fund/nav/history/"+str(i)+"?page=1&size=10000"
    headers= {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.63 Safari/537.36'}
    res = requests.get(url, headers=headers)#遇见阻碍等待15秒,timeout=15
    res.encoding = 'utf-8'
    s = json.loads(res.text)

    s=s['data']['items']
    for y in s:
        y.setdefault("percentage", 'None')
        print("代码："+i+"   时间："+y['date']+"   涨幅："+y['percentage']+"   净值："+y['nav'])
        code_l.append(i)
        time.append(y['date'])
        per.append(y['percentage'])
        nav_l.append(y['nav'])
#转为字典的存储方式，以便于转为DataFrame
for x in time:
    t={"代码":code_l,"时间":time,"涨幅":per,"净值":nav_l}
#将爬取得到的数据转为DataFrame
fund_nav_l=DataFrame(t)
#将基金代码和名称联系起来，按代码填上该基金的名称和类型
fund = pd.merge(q,fund_nav_l, how='inner', on=['代码'])
```
