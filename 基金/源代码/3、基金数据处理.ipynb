```python
import pandas as pd
import numpy as np
from pandas import read_csv
import urllib.request #引入包
import requests
import json
pd.options.display.max_columns = None
```

```python
sj=pd.read_csv(r'./数据/基金所有信息.csv',dtype={'代码':str})#让代码读取数据为str
sj.info()
```

```python
sj
```

```python

#将当天日期为None的行删掉
fund_sj=sj[sj['当天日期']!="None"]
```

```python
#将近一个月涨幅为None的行删掉
fund_sj=fund_sj[fund_sj['近一个月涨幅']!="None"]
fund_sj=fund_sj[fund_sj['近三个月涨幅']!="None"]
fund_sj=fund_sj[fund_sj['近六个月涨幅']!="None"]
fund_sj=fund_sj[fund_sj['今年以来涨幅']!="None"]
```

```python
fund_sj['近一个月涨幅'] = pd.to_numeric(fund_sj['近一个月涨幅']) 
fund_sj['近三个月涨幅'] = pd.to_numeric(fund_sj['近三个月涨幅']) 
fund_sj['近六个月涨幅'] = pd.to_numeric(fund_sj['近六个月涨幅']) 
fund_sj['今年以来涨幅'] = pd.to_numeric(fund_sj['今年以来涨幅']) 
```

```python
fund_sj.info()
```

```python
#为了数据的美观，保存两位小数

fund_sj['近一个月涨幅']= fund_sj['近一个月涨幅'].round(2)
fund_sj['近三个月涨幅']= fund_sj['近三个月涨幅'].round(2)
fund_sj['近六个月涨幅']= fund_sj['近六个月涨幅'].round(2)
fund_sj['今年以来涨幅']= fund_sj['今年以来涨幅'].round(2)
fund_sj.head()
```

```python
fund_sj
```

```python
#将数据保存为csv文件
fund_sj.to_csv('./数据/基金处理后信息.csv',encoding='utf_8_sig')
```
