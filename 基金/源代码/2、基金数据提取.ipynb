```python
import pandas as pd
import numpy as np
from pandas import read_csv
import urllib.request #引入包
import requests
import json
from requests.adapters import HTTPAdapter
```

```python
jijin=pd.read_excel(r'./代码/所有基金代码.xlsx',dtype={'代码':str})#让代码读取数据为str
jijin.info()
```

```python
print(jijin['类型']=='股票型')
```

```python
jijin['代码']
```

```python
jijin['名称']
```

```python
fd_n=[]
end_d=[]#日期
unit_n=[]#净值
nav_g1=[]#一个月涨幅
nav_g3=[]#三个月涨幅
nav_g6=[]#六个月涨幅
nav_gt=[]#今年以来涨幅
s = requests.session()


#max_retries=3 #重试3次
s.mount('http://', HTTPAdapter(max_retries=3))
#遍历基金的代码，以获取每个基金的信息
for r in jijin['代码']:
    url = "https://danjuanapp.com/djapi/fund/derived/"+str(r)+""
    headers= {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.63 Safari/537.36'}
    res = requests.get(url, headers=headers)#遇见阻碍等待15秒,timeout=15
    res.encoding = 'utf-8'
    s = json.loads(res.text)
    #爬取的代码中数据有data的进入循环，没有即跳过
    if 'data' in s.keys() :
        s=s['data']
#         print(s.keys())
        #把没有en_date的数据自动加上。防止报错。爬取出来后再进行数据处理
        s.setdefault("fd_code", 'None')
        s.setdefault("end_date", 'None')
        s.setdefault("nav_grl1m", 'None')
        s.setdefault("nav_grl3m", 'None')
        s.setdefault("nav_grl6m", 'None')
        s.setdefault("nav_grlty", 'None')
        print("代码: "+s['fd_code']+"   当天日期："+s['end_date']+ "   当天净值："+s['unit_nav']+ "    近一个月涨幅："+s['nav_grl1m']+ "    近三个月涨幅："+s['nav_grl3m']+ "    近六个月涨幅："+s['nav_grl6m']+ "    今年以来涨幅："+s['nav_grlty'])
        #将数据加进列表中
        fd_n.append(s['fd_code'])
        end_d.append(s['end_date'])
        unit_n.append(s['unit_nav'])
        nav_g1.append(s['nav_grl1m'])
        nav_g3.append(s['nav_grl3m'])
        nav_g6.append(s['nav_grl6m'])
        nav_gt.append(s['nav_grlty'])
    print("---------------------------------------------")
```

```python
#转为字典的存储方式，以便于转为DataFrame
t={"代码":fd_n,"当天日期":end_d,"当天净值":unit_n,"近一个月涨幅":nav_g1,"近三个月涨幅":nav_g3,"近六个月涨幅":nav_g6,"今年以来涨幅":nav_gt}
t
```

```python
#将爬取得到的数据转为DataFrame
from pandas.core.frame import DataFrame
import pandas as pd
fund_nav=DataFrame(t)
fund_nav
```

```python
#将基金代码和名称的两个csv联系起来，按代码填上该基金的名称和类型
fund_nav = pd.merge(jijin,fund_nav, how='inner', on=['代码'])
```
