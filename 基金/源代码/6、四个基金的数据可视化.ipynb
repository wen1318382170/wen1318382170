## 可视化（由于数据过多，用折线图来展示净值和时间的关系会显得图特别难看，所以为了美化图形，只取四个基金一个月的数据。对四个基金的净值进行分析，国投瑞银进宝灵活配置混合和国投瑞银先进制造混合的净值数据相比接近，金信民兴债券A	和中融中证煤炭指数(LOF)的净值数据相比接近，所以将四个基金分为两个图）

```python
#先对数据进行提取，取近30天的数据
fund1=fund1.head(30)
fund1=fund1.sort_index(ascending=False)#按降序排列
print(fund1)
fund2=fund2.head(30)
fund2=fund2.sort_index(ascending=False)
print(fund2)
```

```python

fund3=fund3.head(30)
fund3=fund3.sort_index(ascending=False)
fund3
fund4=fund4.head(30)
fund4=fund4.sort_index(ascending=False)
fund3
```

## 可视化（折线图）

```python
plt.rcParams['font.sans-serif'] = [u'SimHei']
plt.rcParams['axes.unicode_minus'] = False
fig = plt.figure(figsize=(12,8))

ax = fig.add_subplot(111)
xticks=list(range(0,len(fund1['时间']),3))
ax.set_xticks(xticks)
# yticks=list(range(0,len(fund2['净值']),3))

# ax.set_yticks(yticks)
xticks.append(len(fund1['时间']))
x=fund1['时间']
y2=fund1['净值']
ax.plot(x,y2,marker='o')

ax.plot(fund2['时间'],fund2['净值'],marker='o')#
# 设置轴标签
ax.set_xlabel('时间', fontsize = 15)
# ax.set_ylabel('净值', fontsize = 15)
ax.set_title(' ')
plt.legend(['国投瑞银进宝灵活配置混合', '国投瑞银先进制造混合'])
```
